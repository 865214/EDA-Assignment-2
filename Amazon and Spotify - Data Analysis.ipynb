## 6. What are the most popular product keywords?

```python
from sklearn.feature_extraction.text import CountVectorizer

product_details = data['product_name']

# Initialize CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_features=10)

# Fit and transform the product details
X = vectorizer.fit_transform(product_details)

# Get the feature names (keywords)
keywords = vectorizer.get_feature_names_out()

# Display the top 10 keywords
print(keywords)
```

### The most popular product keywords are Black, Cable, Charging, Compatible, fast, Smar, Tv, Type, usb and White

## 7. What are the most popular product reviews?

```python
top_rated_products = data.groupby('product_details')['rating_count'].sum().nlargest(5)

# Plot the top 5 products by the number of ratings
plt.figure(figsize=(10, 6))
top_rated_products.plot(kind='bar', color='skyblue')
plt.title('Top 5 Products by Number of Ratings')
plt.xlabel('Product Name')
plt.ylabel('Number of Ratings')
plt.xticks(rotation=45)
plt.show()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to get sentiment score
def get_sentiment(text):
    return sia.polarity_scores(text)['compound']

# Apply sentiment analysis to the review text
data['sentiment_score'] = data['review_content'].apply(get_sentiment)

# Classify sentiment into Positive, Negative, or Neutral
data['sentiment_label'] = data['sentiment_score'].apply(
    lambda x: 'Positive' if x > 0.1 else ('Negative' if x < -0.1 else 'Neutral')
)

# Filter positive reviews
positive_reviews = data[data['sentiment_label'] == 'Positive']

# Get the most popular products with positive reviews (based on rating count)
popular_positive_reviews = positive_reviews.groupby('product_name')['rating_count'].sum().nlargest(5)

# Plot the top 5 popular products with positive reviews
plt.figure(figsize=(10, 6))
popular_positive_reviews.plot(kind='bar', color='lightcoral')
plt.title('Top 5 Popular Products with Positive Reviews')
plt.xlabel('Product Name')
plt.ylabel('Number of Ratings')
plt.xticks(rotation=45)
plt.show()
```

## 9.Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.


```python
top_artists = data2.groupby('Artist')['Popularity'].mean().nlargest(10).index
filtered_data = data2[data2['Artist'].isin(top_artists)]

# Set figure size
plt.figure(figsize=(12, 6))

# Create a box plot
sns.boxplot(x='Artist', y='Duration (ms)', data=filtered_data)
plt.xticks(rotation=45)  # Rotate artist names for readability
plt.title("Variation in Track Duration Across Top 10 Artists")
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Artist', y='Duration (ms)', data=filtered_data, inner="quartile")
plt.xticks(rotation=45)
plt.title("Variation in Track Duration Across Top 10 Artists")
plt.xlabel("Artist")
plt.ylabel("Duration (ms)")
plt.show()
```

### Yes the duration of track varries between the top artist because IQR are differnce in height and Kendric lemar artist having out lier in their duration so there is an difference
